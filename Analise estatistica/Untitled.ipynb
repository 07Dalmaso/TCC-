```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Carrega os dados do arquivo CSV
file_path_resultados = 'resultado_2019.csv'  # atualize com o caminho real
df = pd.read_csv(file_path_resultados)

# Remove as colunas 'COD_CIDADE' e 'COD_REG' do DataFrame
df_relevante = df.drop(columns=['COD_CIDADE', 'COD_REG'])

# Análise descritiva de todas as linhas
print("Análise descritiva:")
print(df_relevante.describe())

# Seleciona apenas colunas numéricas para a correlação
numerical_df = df_relevante.select_dtypes(include=[np.number])

# Verificando a correlação entre as variáveis
correlation_matrix = numerical_df.corr()
print("\nMatriz de correlação:")
print(correlation_matrix)

# Visualizando a matriz de correlação
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Matriz de Correlação')
plt.show()

# Scatter plots entre 'Casos' e cada feature
features = [
    'Média de Precipitação Normalizada',
    'Média de Temperatura Máxima Normalizada',
    'Média de Temperatura Mínima Normalizada',
    'Média de Umidade Relativa Máxima Normalizada',
    'Média de Umidade Relativa Mínima Normalizada',
    'População Normalizada'
]

for feature in features:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=df_relevante, x=feature, y='Casos')
    plt.title(f'Casos de Dengue vs. {feature}')
    plt.xlabel(feature)
    plt.ylabel('Casos de Dengue')
    plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Carrega os dados do arquivo CSV
file_path_resultados = 'resultado_2019.csv'  # atualize com o caminho real
df = pd.read_csv(file_path_resultados)

# Remove as colunas 'COD_CIDADE' e 'COD_REG' do DataFrame
df_relevante = df.drop(columns=['COD_CIDADE', 'COD_REG'])

# Análise descritiva de todas as linhas
print("Análise descritiva:")
print(df_relevante.describe())

# Seleciona apenas colunas numéricas para a correlação
numerical_df = df_relevante.select_dtypes(include=[np.number])

# Verificando a correlação entre as variáveis
correlation_matrix = numerical_df.corr()
print("\nMatriz de correlação:")
print(correlation_matrix)

# Visualizando a matriz de correlação
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Matriz de Correlação')
plt.show()

# Separa os dados do Rio de Janeiro e do restante do DF
# Considerando os últimos 4 registros como sendo do Rio de Janeiro
df_rio = df_relevante.iloc[-4:]  # Últimos 4 registros
df_df = df_relevante.iloc[:-4]    # Todos os registros menos os últimos 4

# Lista de features para comparação
features = [
    'Média de Precipitação Normalizada',
    'Média de Temperatura Máxima Normalizada',
    'Média de Temperatura Mínima Normalizada',
    'Média de Umidade Relativa Máxima Normalizada',
    'Média de Umidade Relativa Mínima Normalizada',
    'População Normalizada'
]

# Criando gráficos de dispersão para cada feature
for feature in features:
    plt.figure(figsize=(10, 6))
    
    # Scatter plot para DF
    sns.scatterplot(data=df_df, x=feature, y='Casos', label='Distrito Federal', color='blue', alpha=0.6)
    
    # Scatter plot para Rio de Janeiro
    sns.scatterplot(data=df_rio, x=feature, y='Casos', label='Rio de Janeiro', color='red', alpha=0.6)
    
    plt.title(f'Casos de Dengue vs. {feature}')
    plt.xlabel(feature)
    plt.ylabel('Casos de Dengue')
    plt.legend()
    plt.show()
```
